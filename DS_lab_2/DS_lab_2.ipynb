```python
#все вище - спроби очистити дані
#на жаль цьому нас не вчили PoroSad  
#-----------------------------------------------------------------
data = pd.read_excel("rollingsales_manhattan.xls",parse_dates = True, verbose=True, na_values=[' ','  ','   ','\t','\n','\n\n',' \n'])#тому буду чистити частково, наскільки зможу
data
```

```python
data.isnull().sum()
```

```python
#невід'ємна ціна
cleared_data = data[data["SALE PRICE"]>0]
#виключаю без району, умова виглядає як костиль
cleared_data = cleared_data[cleared_data["NEIGHBORHOOD"].str.isspace() != True]
#видаляю непотрібні стовпці
cleared_data = cleared_data.drop(columns=["EASE-MENT","LOT","BLOCK","APARTMENT NUMBER","ZIP CODE"])
cleared_data
```

```python
#ціна продажу за 12 рік відносно 13-го
cleared_data.groupby(cleared_data["SALE DATE"].dt.year).sum().plot(kind="pie", y="SALE PRICE", autopct='%.1f')
```

```python
cleared_data.groupby([cleared_data["SALE DATE"].dt.year, cleared_data["SALE DATE"].dt.month]).sum()#10 нулів
```

```python
cleared_data.groupby([cleared_data["SALE DATE"].dt.year, cleared_data["SALE DATE"].dt.month]).sum().plot(
    kind="bar", y="SALE PRICE",legend="",xlabel="Sale date by month",ylabel="Sum sale price\n1:10^10$")
```
